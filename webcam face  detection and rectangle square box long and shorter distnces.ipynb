```python
import cv2
import dlib
import face_recognition
import os
```

```python
webcam_video=cv2.VideoCapture(0)
number_of_faces=[]

while True:
    ret,current_frame=webcam_video.read()
    number_of_faces=face_recognition.face_locations(current_frame,model="hog")
    print(number_of_faces)
    
    for index,current_face_location in enumerate(number_of_faces):
        top_pos,right_pos,bottom_pos,left_pos=current_face_location
        current_image=current_frame[top_pos:bottom_pos,left_pos:right_pos]
        cv2.imshow("Face no"+str(index+1),current_image)
    
    if cv2.waitKey(1) & 0xff == ord("q"):
        break

    
webcam_video.release()
cv2.destroyAllWindows()       
```

# Web cam with rectangle square box face detection for shorter to webcam 

```python
webcam_video=cv2.VideoCapture(0)
number_of_faces=[]

while True:
    ret,current_frame=webcam_video.read()
    resize_frame=cv2.resize(current_frame,(0,0),fx=0.25,fy=0.25)
    number_of_faces=face_recognition.face_locations(resize_frame,model="hog")
    print(number_of_faces)
    
    for index,current_face_location in enumerate(number_of_faces):
        top_pos,right_pos,bottom_pos,left_pos=current_face_location
        top_pos=top_pos*4
        right_pos=right_pos*4
        bottom_pos=bottom_pos*4
        left_pos=left_pos*4
        cv2.rectangle(current_frame,(left_pos,top_pos),(right_pos,bottom_pos),(0,0,255),2)
    cv2.imshow("Web cam video",current_frame)
    if cv2.waitKey(1) & 0xff == ord("q"):
        break

    
webcam_video.release()
cv2.destroyAllWindows()      
```

# Web cam with rectangle square box face detection for long to webcam aslo (use the parameter number of times to upsamplein below code )

```python
webcam_video=cv2.VideoCapture(0)
number_of_faces=[]

while True:
    ret,current_frame=webcam_video.read()
    resize_frame=cv2.resize(current_frame,(0,0),fx=0.25,fy=0.25)
    number_of_faces=face_recognition.face_locations(resize_frame,number_of_times_to_upsample=5,model="hog")
    print(number_of_faces)
    
    for index,current_face_location in enumerate(number_of_faces):
        top_pos,right_pos,bottom_pos,left_pos=current_face_location
        top_pos=top_pos*4
        right_pos=right_pos*4
        bottom_pos=bottom_pos*4
        left_pos=left_pos*4
        cv2.rectangle(current_frame,(left_pos,top_pos),(right_pos,bottom_pos),(0,0,255),2)
    cv2.imshow("Web cam video",current_frame)
    if cv2.waitKey(1) & 0xff == ord("q"):
        break

    
webcam_video.release()
cv2.destroyAllWindows()
```
